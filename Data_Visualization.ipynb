```python
# Pisahkan data berdasarkan tahun
data["Year"] = data["Date"].dt.year
data_2018 = data[data["Year"] == 2018]
data_2019 = data[data["Year"] == 2019]
```

```python
# Plot Line Chart untuk pengembalian kumulatif
data["Cumulative Return"] = (1 + data["Return"] / 100).cumprod()
plt.figure(figsize=(14, 7))
sns.lineplot(data=data, x="Date", y="Cumulative Return", hue="Year", palette="Set1", marker="o")
plt.title("Pengembalian Yang Lebih Baik Pemegang Bitcoin (2018 vs 2019)", fontsize=14)
plt.ylabel("Pengembalian Kumulatif", fontsize=12)
plt.xlabel("Minggu", fontsize=12)
plt.legend(title="Tahun", loc="upper left")
plt.grid(axis="both", linestyle="--", alpha=0.7)
plt.show()
```

Soal 2 : Permen

Terdapat sekantong permen dengan lima jenis permen. Buat diagram yang menunjukkan persentase peluang bahwa kita akan mengeluarkan permen Snickers dari kantong jika kita melakukan pengambilan acak. Sebutkan peluang memilih permen Snickers.

```python
# Data
candy_names = ['Kit Kat', 'Snickers', 'Milky Way', 'Toblerone', 'Twix']
candy_counts = [52, 39, 78, 13, 78]
```

```python
# Hitung total permen dan peluang Snickers
total_candies = sum(candy_counts)
snickers_count = candy_counts[candy_names.index("Snickers")]
snickers_probability = snickers_count / total_candies
```

```python
# Cetak peluang Snickers
print(f"Peluang memilih Snickers: {snickers_probability:.2%}")
```

```python
# Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(
    candy_counts,
    labels=candy_names,
    autopct='%1.1f%%',
    startangle=140,
    colors=["#FFC300", "#FF5733", "#C70039", "#900C3F", "#581845"],
    explode=[0, 0.1, 0, 0, 0]  # Highlight Snickers
)
plt.title("Distribusi Permen dalam Kantong", fontsize=14)
plt.show()
```

1.Chart apa yang anda pilih untuk problem diatas dan mengapa anda memilih chart tersebut?

Karena Pie chart sangat baik untuk menunjukkan proporsi relatif setiap elemen terhadap total. Dalam kasus ini, kita dapat melihat bagaimana permen Snickers dibandingkan dengan jenis permen lainnya di dalam kantong.

2.Berapa persen kemungkinan Anda akan memilih Snickers saat mengeluarkan permen dari tas secara acak?

Soal 3: Makanan

Restoran memiliki daftar menu makanan penutup yang terlalu banyak. Mereka ingin menghilangkan beberapa item dari menu. Untuk membuat sebagian besar pelanggan mereka senang, mereka ingin menghapus hanya tiga makanan penutup yang paling tidak populer. Kita memiliki daftar makanan penutup yang disajikan restoran, serta hitungan berapa kali makanan penutup tersebut dijual dalam seminggu terakhir. Buatlah visualisasi yang menunjukkan popularitas relatif dari makanan penutup. Sebutkan tiga makanan penutup yang harus disingkirkan.

```python
# Data
dessert_sales = {
    'Lava Cake': 14,
    'Mousse': 5,
    'Chocolate Cake': 12,
    'Ice Cream': 19,
    'Truffles': 6,
    'Brownie': 8,
    'Chocolate Chip Cookie': 12,
    'Chocolate Pudding': 9,
    'Souffle': 10,
    'Chocolate Cheesecake': 17,
    'Chocolate Chips': 2,
    'Fudge': 9,
    'Mochi': 13,
}
```

```python
# Buat DataFrame
data = pd.DataFrame(list(dessert_sales.items()), columns=["Dessert", "Sales"])
data = data.sort_values(by="Sales", ascending=True)
```

```python
# Identifikasi tiga makanan penutup dengan penjualan terendah
least_popular = data.head(3)
```

```python
# Cetak tiga makanan penutup yang harus dihapus
print("Tiga makanan penutup yang harus disingkirkan:")
print(least_popular)
```

```python
# Visualisasi Bar Chart
plt.figure(figsize=(10, 6))
sns.barplot(data=data, x="Sales", y="Dessert", palette="viridis")
plt.title("Popularitas Makanan Penutup", fontsize=14)
plt.xlabel("Penjualan (unit)", fontsize=12)
plt.ylabel("Makanan Penutup", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.axvline(least_popular["Sales"].max(), color="red", linestyle="--", label="Ambang Batas")
plt.legend()
plt.tight_layout()
plt.show()
```
